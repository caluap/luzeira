```python
delta = {
  'AR':   0, # aries
  'TA':  30, # taurus
  'GE':  60, # gemini
  'CN':  90, # cancer
  'LE': 120, # leo
  'VI': 150, # virgo
  'LI': 180, # libra
  'SC': 210, # scorpio
  'SG': 240, # sagittarius
  'SA': 240, # sagittarius (there are some inconsistencies in the data!)
  'CP': 270, # capricorn
  'AQ': 300, # aquarius
  'PI': 330  # pisces
}
```

```python
import json
from pprint import pprint


ephemerides = {}

for i_year in range(1900, 2101):
  s_year = str(i_year)
  with open('raw_data/' + s_year + '.json') as f:
    year_data = json.load(f)
    
  ephemerides[s_year] = {}
  print(f'year:{i_year}...')

  for i_month, month in year_data.items():
    ephemerides[s_year][i_month] = {}
#       print(f'   month:{i_month}...')

    for i_day, day in month.items():
      ephemerides[s_year][i_month][i_day] = {}
#         print(f'      day:{i_day}...')

      for key, value in day.items():
        if key == 'time':
          continue
        angle = int(value[0:2])
        angle += delta[value[2:4]]
        angle += int(value[4:])/60

        ephemerides[s_year][i_month][i_day][key] = angle
          
        
  fn = 'clean_data/' + s_year + '.json'
  with open(fn, 'w') as fp:
    json.dump(ephemerides[s_year], fp, sort_keys=True, indent=2)
```
